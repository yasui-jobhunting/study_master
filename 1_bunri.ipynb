```python
#検出テキストからフレームごとcsvに関数
import cv2
import csv
import numpy as np
import math
import os
from PIL import Image
import glob

data_list=[]
data_left=[]
data_right=[]
error=[]

totle_count=0
totalframecount =0

def save_frame(video_name,frame_num,result_path):  #画像を保存する　#result_path:切り出した画像の保存先

    # video_path="./data/video/{}.mp4".format(video_name)
    video_path=video_name
    cap = cv2.VideoCapture(video_path)  #動画の読み込み

    if not cap.isOpened():
        return


    os.makedirs(os.path.dirname(result_path), exist_ok=True)

    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)

    ret, frame = cap.read()     #フレームを取得

    if ret:
        cv2.imwrite(result_path, frame)  



def get_lRGB(image_path,i,data_num): #画像中のRGB値を取得する(左)

    # frame=cv2.imread(image_path)
    image=cv2.imread(image_path)    #image[y,x]
    
    x=int(data_left[data_num-1][0])
    y=int(data_left[data_num-1][1])

    w=int(data_left[data_num-1][2])
    h=int(data_left[data_num-1][3])

    h_coo,w_coo = int(h*0.8) , int(w*0.8) #Jul 20
    
    xx=int(x+w*0.1) #Jul 20
    yy=int(y+h*0.1)   #Jul 20

    avr = [0,0,0]
    b_ave,g_ave,r_ave=0,0,0
    # image = frame[left+h_coo,top+w]
    ll=0

    # print(image)
    # print(data)
    # print(image_path,xx,xx+w,yy,yy+h)
    if yy+h_coo>1080:
        h_coo=1080-yy
       
    if xx+w_coo>1920:
        w_coo=1920-xx

    for iii in range(yy,yy+h_coo):
        for j in range(xx,xx+w_coo):
            #画素値[0,0,0]（Black）を除外してピクセルの和とbgrの画素値の合計を計算する
            if int(image[iii,j,0])+int(image[iii,j,1])+int(image[iii,j,2])>200 :
                ll+=1    #対象となるピクセル数を計算する
                #対象となるピクセルの画素値の和を計算する
                b_ave=b_ave+image[iii,j,0]
                g_ave=g_ave+image[iii,j,1]
                r_ave=r_ave+image[iii,j,2]

    #RGB値が全部0だったとき、閾値処理をしない
    if r_ave==0 and g_ave==0 and b_ave==0:
        for iii in range(yy,yy+h_coo):
            for j in range(xx,xx+w_coo):
                #画素値[0,0,0]（Black）を除外してピクセルの和とbgrの画素値の合計を計算する
                # if int(image[iii,j,0])+int(image[iii,j,1])+int(image[iii,j,2])>200 :
                ll+=1    #対象となるピクセル数を計算する
                #対象となるピクセルの画素値の和を計算する
                b_ave=b_ave+image[iii,j,0]
                g_ave=g_ave+image[iii,j,1]
                r_ave=r_ave+image[iii,j,2]

        #画素値合計をピクセル数で除することでRGBの画素値の平均値を求める
    # if ll>100:
    if ll>0:
        b_ave=b_ave/ll
        g_ave=g_ave/ll
        r_ave=r_ave/ll
    
    

    avr=[r_ave, g_ave, b_ave]

    data_left.append([x,y,h,w,avr[0],avr[1],avr[2]])

def get_rRGB(image_path,i,data_num): #画像中のRGB値を取得する(右)

    # frame=cv2.imread(image_path)
    image=cv2.imread(image_path)    #image[y,x]

    x=int(data_right[data_num-1][0])
    y=int(data_right[data_num-1][1])

    w=int(data_right[data_num-1][2])
    h=int(data_right[data_num-1][3])

    h_coo,w_coo = int(h*0.8) , int(w*0.8) #Jul 20
    
    xx=int(x+w*0.1) #Jul 20
    yy=int(y+h*0.1)   #Jul 20

    avr = [0,0,0]
    b_ave,g_ave,r_ave=0,0,0
    # image = frame[left+h_coo,top+w]
    ll=0

    # print(image)
    # print(data)
    # print(image_path,xx,xx+w,yy,yy+h)
    if yy+h_coo>1080:
        h_coo=1080-yy
       
    if xx+w_coo>1920:
        w_coo=1920-xx

    for iii in range(yy,yy+h_coo):
        for j in range(xx,xx+w_coo):
            #画素値[0,0,0]（Black）を除外してピクセルの和とbgrの画素値の合計を計算する
            if int(image[iii,j,0])+int(image[iii,j,1])+int(image[iii,j,2])>200 :
                ll+=1    #対象となるピクセル数を計算する
                #対象となるピクセルの画素値の和を計算する
                b_ave=b_ave+image[iii,j,0]
                g_ave=g_ave+image[iii,j,1]
                r_ave=r_ave+image[iii,j,2]

    #RGB値が全部0だったとき、閾値処理をしない
    if r_ave==0 and g_ave==0 and b_ave==0:
        for iii in range(yy,yy+h_coo):
            for j in range(xx,xx+w_coo):
                #画素値[0,0,0]（Black）を除外してピクセルの和とbgrの画素値の合計を計算する
                # if int(image[iii,j,0])+int(image[iii,j,1])+int(image[iii,j,2])>200 :
                ll+=1    #対象となるピクセル数を計算する
                #対象となるピクセルの画素値の和を計算する
                b_ave=b_ave+image[iii,j,0]
                g_ave=g_ave+image[iii,j,1]
                r_ave=r_ave+image[iii,j,2]

        #画素値合計をピクセル数で除することでRGBの画素値の平均値を求める
    # if ll>100:
    if ll>0:
        b_ave=b_ave/ll
        g_ave=g_ave/ll
        r_ave=r_ave/ll
    
    avr=[r_ave, g_ave, b_ave]

    data_right.append([x,y,h,w,avr[0],avr[1],avr[2]])

# save_frame('data/temp/sample_video.mp4', 100, 'data/temp/result_single/sample_100.jpg')

def get_imageRGB(video_name,frame_num,result_path,i,muki,*data): #特定の画像のRGB値を取得する
    if frame_num<totalframecount:
        if not os.path.isdir(result_path):
            save_frame(video_name,frame_num,result_path) #画像を保存
        if muki==True:
            get_lRGB(result_path,i,frame_num) #保存した画像からRGB値を取得
        else:
            get_rRGB(result_path,i,frame_num)
```
